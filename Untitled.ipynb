```python
import pandas as pd
```

```python
df = pd.read_csv("nba.csv.txt")
df
```

```python
#Average Points Scored
df["points"].mean()
```

```python
#Highest Scorer
print("Top Scorer:")
df[df["points"] == df["points"].max()]
```

```python
#Players that Scored at Least 25 Points
df[df["points"]>25]
```

```python
#Average Player Points per Team
print("Average Points per Team:")
team_avg = df.groupby("team")["points"].mean()
team_avg
```

```python
#Team Rank by Scoring
print("Team Rank:")
team_avg.sort_values(ascending = False)
```

```python
print("Top Insight: Teams show small variation in average player scoring, with slight differences between rosters.")
```
