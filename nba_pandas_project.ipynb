```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nba.csv.txt")
df
```

## Overview

```python
df.head()
```

```python
print("\nDataset Information:")
df.info()
```

## Basic Stats

```python
print("\nAverage Points Scored:")
average_points = df["points"].mean()
average_points
```

## Player Analysis

```python
print("\nTop Scorer:")
top_scorer = df[df["points"] == df["points"].max()]
top_scorer
```

```python
print("\nTop 5 Scorers:")
top5 = df.sort_values(by = "points", ascending = False).head()
top5
```

```python
print("\nPlayers Scoring More Than 25 Points:")
high_scores = df[df["points"]>25]
high_scores
```

## Team Analysis

```python
print("\nAverage Points per Team:")
team_avg = df.groupby("team")["points"].mean()
team_avg
```

## Team Ranking

```python
print("\nTeam Ranking by Average Points:")
team_avg.sort_values(ascending = False)
```

## Visualization

```python
team_avg.sort_values().plot(kind="barh")

plt.title("Average Points Per Team")
plt.xlabel("Average Points")
plt.ylabel("Team")

plt.show()
```

## Assists Analysis

```python
print("\nAverage Assists Per Team")
df.groupby("team")["assists"].mean()
```

## Conclusion

```python
print("\nTop Insight: Teams show relatively small variation in average player scoring, suggesting balanced offensive production across many rosters.")
```
